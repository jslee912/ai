# 2. LLM 답변 생성
## 2.1 OpenAI SDK를 선택
- pip install openai / pip install langchain-openai

```python
from openai import OpenAI
client = OpenAI()
response  = client.chat.completions.create(
    model = "gpt-4o-mini",
    messages=[
        {"role":"system", "content":"너는 친절하게 대답해 주는 비서야"},
        {"role":"user", "content":"2020년 월드 시리즈 누가 우승했어?"}
    ],
    temperature=0.5, # 0~0.3:일관되고 예측가능한답 / 0.8~1:창의적다양한답변
)
```

```python
response.choices[0].message.content
```

```python
response.usage  # 사용 토큰수  completion_tokens=68(출력), prompt_tokens=38(입력)
```

## 2.2 LangChain을 선택

```python
from langchain_openai import ChatOpenAI
# .env에 OPENAI_API_KEY 등록되어있다는 전제하에
llm = ChatOpenAI(
    model = "gpt-4o-mini",
    temperature=0.5
)
ai_message = llm.invoke("2020년 월드 시리즈는 누가 우승했어요?")  # llm.invoke()과금
```

```python
ai_message.content
```

```python
ai_message.usage_metadata
```

```python
from langchain_core.messages import SystemMessage, HumanMessage
messages = [
    SystemMessage(content="너는 친절하게 대답해 주는 비서야"),
    HumanMessage(content="2020년 월드 시리즈 누가 우승했어?")
]
ai_message = llm.invoke(messages)
print(ai_message.content)
```
